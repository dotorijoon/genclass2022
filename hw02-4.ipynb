```python
# 영상의 의미지를 연속적으로 캡쳐할 수 있게 하는 class
vidcap = cv2.VideoCapture('public/hw02/remastering.mp4')
 
count = 0 
while(vidcap.isOpened()):
    ret, image = vidcap.read() 
    # 캡쳐된 이미지를 저장하는 함수
    try:
        cv2.imwrite("public/hw02/frames/frame%d.png" % count, image) 
        #print('Saved frame%d.png' % count)
    except:
        break
    count += 1
 
vidcap.release()
```

```python
# 변경된 이미지를 하나의 list로 
frames = []
for icnt in range(1629):
    ioarray = imageio.imread(f'GFPGAN/results/restored_imgs/frame{icnt}.png')
    frames.append(ioarray)
    #len(frames)
len(frames)    
```

```python
# 이미지를 Data를 Imageio로 동영상으로 변환
def generate_mp4(out_name, images, kwargs):
    writer = imageio.get_writer(out_name , **kwargs)
    for image in images:
        writer.append_data(image)
    writer.close()

generate_mp4('remastered.mp4', frames, {'fps':30})
```
