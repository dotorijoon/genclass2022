```python
import IPython.display as ipd
import imageio
import os
import cv2
from PIL import Image
```

```python
def generate_mp4(out_name, images, kwargs):
    writer = imageio.get_writer(out_name , **kwargs)
    for image in images:
        writer.append_data(image)
    writer.close()
```

```python
frames = []
for icnt in range(1,241):
    ioarray = imageio.imread('public/hw03/{0:06d}.png'.format(icnt))
    frames.append(ioarray)
    
generate_mp4('input.mp4', frames, {'fps':24})
re  = ipd.Video('input.mp4',width=512, height=288)
display(re)
```

```python
frames_mask = []
for icnt in range(1,241):
    ioarray = imageio.imread('public/hw03/{0:06d}_mask.png'.format(icnt))
    frames_mask.append(ioarray)
    
generate_mp4('input_mask.mp4', frames_mask, {'fps':24})
re  = ipd.Video('input_mask.mp4',width=512, height=288)
display(re)
```
